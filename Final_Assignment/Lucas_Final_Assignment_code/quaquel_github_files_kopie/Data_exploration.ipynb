# Setting boundaries and expectations
We think that livelihood and casaulties are most important so we'll take the lowest 20% of those values into consideration as candidate solutions.

```python
candidates = results.nsmallest(328, ['A.2_Expected Number of Deaths', 'A.1_Expected Number of Deaths'])
candidates
```

This gives us an upper boundary of:
1. `'A.1_Expected Number of Deaths'` = 0 
2. `'A.2_Expected Number of Deaths'` = 0.001487

Now that we have defined a new solution set of 328 candidate solutions based on our best 20% performing scenario's we can reevaluate the costs attatched to this.
